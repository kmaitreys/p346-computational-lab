```python
n = int(input("Enter a positive integer greater than 1: "))

s = 0 #declaring the leading number of pattern
for i in range(n):
    s = s + i
    
list = []
def averageDistance(n):
    m = 0
    m = m + s
    for j in range(1, n): #iterating over the pattern n times and appending each output into a list
        m = m - (n - j) + j
        list.append(m)
    list.append(s) 
    x = 0
    for k in range(0, len(list)): #adding all the outputs stored in the list to get the final sum
        x = x + list[k]
    y = x/(n**2) #calculating the average
    print("The average distance between two points on a straight line made of " + str(n) + " discrete points is " + str(y))
        
averageDistance(n)
```
